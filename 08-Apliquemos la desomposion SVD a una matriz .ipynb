```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')

imagen = Image.open("imagenes/imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
```

```python
imagen_grey = imagen.convert('LA') # convertimos a escala de grises 

imagen_matriz = np.array(list(imagen_grey.getdata(band=0)), float)

imagen_matriz.shape = (imagen_grey.size[1], imagen_grey.size[0]) #convertimos imagen vector a una matriz
print(imagen_matriz)
```

```python
U, D, V = np.linalg.svd(imagen_matriz)
```

```python
print(imagen_matriz.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_reconstruida = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])

plt.imshow(imagen_reconstruida, cmap='gray')
```
