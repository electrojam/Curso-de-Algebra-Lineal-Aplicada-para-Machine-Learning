```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# creamos muestra con semilla seed

np.random.seed(42)

# cramos nuestra variable x

x = 3*np.random.rand(200)
y = 20*x + 2*np.random.rand(200)

# damos forma a nuestros vectores

x = x.reshape(200, 1)
y = y.reshape(200, 1)

print(x)
print(y)
```

```python
# creamos un vector con pares unidos de x y

xy = np.hstack([x, y])
print(xy.shape)
```

```python
plt.plot(xy[:, 0], xy[:, 1], '.')
plt.show()

# la gr;áfica muestra un conjunto de datos altamente correlacionados
```

```python
# la gr;áfica muestra un conjunto de datos altamente correlacionados
# podemos eliminar una de las dimensiones si logramos escribir los puntos de otra manera
# ¿cuanta informaci;ón estari;ámos perdiendo en caso de quedarnos con una sola variable?
# ¿como podemos girar nuestro sistema de referencia para que nuestra informaci;ón?... 
# quede representada sobre cada uno de los ejes y eligir con cual eje nos quedamos

xy_centrado = xy - np.mean(xy, axis = 0)

plt.plot(xy_centrado[:, 0], xy_centrado[:, 1], '.')
plt.show()

# ahora la gr;áfica se muestra centrado (en la mitad) al rededor del 0
```

```python
# buscaremos las componentes principales para resolver un problema

# calculamos d

autovalores, autovectores = np.linalg.eig(xy_centrado.T.dot(xy_centrado))
print(autovectores)

# los autovectores son los que maximizan nuestra funci;ón
```
